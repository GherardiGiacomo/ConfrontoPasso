<h3>Esplorazione dei dati dell'accelerometro: confronto tra camminate con e senza plantari </h3> <br>

<h1> Obiettivi </h1> <br>
Lo scopo di questo progetto è quello di confrontare i dati acquisiti <i>( tramite l'applicazione <b>Sensor Logger</b>) </i>da diversi percorsi della stessa durata (~ 2 minuti), con e senza l'utilizzo di plantari ortopedici .<br> In particolare,l'analisi è stata effettuata sul sensore accelerometro, per esaminare le variazioni nell'accelerazione durante le attività. 
<br><br><br>Gli obiettivi principali che mi sono posto sono:<br>
    <ul> <li> Comparare le caratteristiche dell'accelerazione registrate durante la camminata su tapis roulant con e senza l'uso di plantari ortopedici </li> 
     <li> Comparare le caratteristiche dell'accelerazione registrate durante la salita e la discesa delle scale,  con e senza l'uso di plantari ortopedici </li>  
 <li> Ridurre il rumore dei dati rilevati tramite l'uso di un filtro (passa-basso) </li>  <li> Esplorare e confrontare eventuali differenze rilevate </li>

<h1> Teoria e metodi </h1>

Il sensore utilizzato per questa analisi è l'accelerometro.

Ho applicato varie tecniche di analisi dati, tra cui calcoli delle medie, identificazione di picchi e valli e l'applicazione di un filto passa-basso.
    
Per esplorare e visualizzare i dati raccolti, ho importato tramite la libreria <b> pandas</b> i file .csv acquisiti da analizzare (convertiti in timestamp). Ho utilizzato inoltre Plotly per generare plot dinamici ed interattivi dei dati

<ul> <li> <h3> SPIEGAZIONE E MOTIVAZIONE DELL'IDENTIFICAZIONE DI PICCHI E VALLI </h3> </li> <BR> <BR>

Tramite la funzione <b>find_peaks</b>, ho prima individuato i picchi positivi nell'accelerazione. Ho analogamente in seguito individuato le valli <i>(questa volta negando il segnale)</i>. La distanza minima tra i picchi è stata impostata a 25 campionamenti per evitare di rilevare picchi troppo vicini tra loro come separati <i>(ragionamento identico applicato poi anche per le valli)</i>. 
Tramite la funzione Plotly ho reso possibile la visualizzazione grafica come plot.

Sono stati infine calcolati i valori medi di picchi <i>(e valli)</i>, al fine di fornire una stima del comportamento medio durante il periodo di rilevamento dei picchi e delle valli. 

<ul> <li> <h3> SPIEGAZIONE E MOTIVAZIONE DEL CALCOLO DELLE FREQUENZA MEDIE </h3> </li> <BR> <BR>

Al fine di garantire coerenza nell'analisi dei dati, ho deciso di calcolare una frequenza di campionamento media.
La funzione <b> df_acce_with['time'].diff()</b> calcola la differenza temporale tra le misurazioni consecutive, fornendo gli intervalli di campionamento, prima per i dati raccolti con i plantari e dopo quelli senza.

Il calcolo della variabile <b>sampling_frequency</b> <i>(ed ovviamente il ragionamento si applica poi per la variabile <b> sampling_frequency_without </b>)</i> mi permette di ottenere la frequenza di campionamento desiderata.

<ul> <li> <h3> TEORIA, SPIEGAZIONE E MOTIVAZIONE DELL'UTILIZZO DEL FILTRO BUTTERWORTH  </h3> </li> <BR> <BR>

Allo scopo di ridurre il rumore nei dati raccolti, ho applicato un filtro passa-basso; nello specifico il filtro <b>Butterworth</b>. 

<img src="Filtro Butterworth.png" alt="Filtro Butterworth" width="350" height="450"/>

Per quanto riguarda i parametri del filtro <b>butter_lowpass</b>, troviamo "<b>cutoff_frequency</b>" che rappresenta la frequenza di taglio del filtro e "<b>order</b>", che rappresenta l'ordine del filtro Butterworth.

Reputo i valori scelti per questi due parametri adeguati ad una pulizia sufficente da non cambiare in modo significativo il segnale

<ul> <li> <h3> SPIEGAZIONE E MOTIVAZIONE DEL CALCOLO MEDIE E DIFFERENZE CON SEGNO</h3> </li> 

Per comprendere meglio le caratteristiche del segnale, ho pensato di osservare come l'accelerazione varia lungo l'asse x rispetto alla media, sia con che senza l'uso dei plantari.

Ho applicato lo stesso ragionamento anche sul segnale filtrato.

Tramite le funzioni <b> mean_x_with = df_acce_with['x'].mean() </b> e <b> mean_x_without = df_acce_without['x'].mean() </b> ho calcolato le medie dei dati lungo l'asse x, mentre con le due seguenti sottrazioni <i> (<b> diff_with = df_acce_with['x'] - mean_x_with </b> e <b> diff_without = df_acce_without['x'] - mean_x_without</b>)</i>, vengono calcolate le differenze con segno rispetto alla media per i dati con e senza l'uso di plantari, in modo tale da valutare quanto ogni campione si discosta dalla media.

<h1> Esperimenti </h1>

```python
import os
from scipy.io import loadmat
import math
import numpy as np
import pandas as pd
from scipy.signal import find_peaks
import plotly.graph_objects as go
import matplotlib.pyplot as plt
from scipy.signal import butter,filtfilt, lfilter,group_delay
from scipy import signal
from scipy.ndimage import convolve1d
```

<h3> Eseguo dei plot per quanto riguarda le due camminate su tapis roulant su una finestra di tempo delimitata da start_index ed end_index, in modo tale da poter vedere graficamente i dati da analizzare. <br> <br>
Eseguo poi un altro plot per quanto riguarda le due camminate sulle scale </h3> <br>
<h4> Accelerometro <br>
Asse x: Rappresenta l'accelerazione lungo l'asse orizzontale (lato a lato). <br>
Asse y: Rappresenta l'accelerazione lungo l'asse verticale (in avanti e indietro). <br>
Asse z: Rappresenta l'accelerazione lungo l'asse laterale (in su e giù). </h4> </h3>

```python
start_index = 400  # Indice iniziale
end_index = 1000    # Indice finale

# Camminata con plantari
df_acce_with = pd.read_csv('tapis_roulant_con_plantari_-2023-11-29_20-37-14/Accelerometer.csv')
df_acce_with.index = pd.to_datetime(df_acce_with['time'], unit='ns')

fig_with = go.Figure()
for axis in ['x', 'y', 'z']:
    selected_data_with = df_acce_with[start_index:end_index]
    fig_with.add_trace(go.Scatter(x=selected_data_with.index, y=selected_data_with[axis], name=axis))
    fig_with.update_layout(title='Camminata su tapis roulant con plantari')

fig_with.show()

# Camminata senza plantari
df_acce_without = pd.read_csv('tapis_roulant_senza_plantari_-2023-11-29_20-41-36/Accelerometer.csv')
df_acce_without.index = pd.to_datetime(df_acce_without['time'], unit='ns')

fig_without = go.Figure()
for axis in ['x', 'y', 'z']:
    selected_data_without = df_acce_without[start_index:end_index]
    fig_without.add_trace(go.Scatter(x=selected_data_without.index, y=selected_data_without[axis], name=axis))
    fig_without.update_layout(title='Camminata su tapis roulant senza plantari')

fig_without.show()
```

```python
# Scale con plantari
df_acce_with = pd.read_csv('scale_giu_-_su_con_plantari_-2023-11-29_20-47-39/Accelerometer.csv')
df_acce_with.index = pd.to_datetime(df_acce_with['time'], unit='ns')

fig_with = go.Figure()
for axis in ['x', 'y', 'z']:
    selected_data_with = df_acce_with[start_index:end_index]
    fig_with.add_trace(go.Scatter(x=selected_data_with.index, y=selected_data_with[axis], name=axis))
    fig_with.update_layout(title='Scale con plantari')

fig_with.show()

# Scale senza plantari
df_acce_without = pd.read_csv('scale_giu_-_su_senza_plantari_-2023-11-29_20-45-38/Accelerometer.csv')
df_acce_without.index = pd.to_datetime(df_acce_without['time'], unit='ns')

fig_without = go.Figure()
for axis in ['x', 'y', 'z']:
    selected_data_without = df_acce_without[start_index:end_index]
    fig_without.add_trace(go.Scatter(x=selected_data_without.index, y=selected_data_without[axis], name=axis))
    fig_without.update_layout(title='Scale senza plantari')

fig_without.show()
```

<h3> Calcolo le valli e i picchi della camminata con plantari  </h3>

```python
signal = df_acce_with['x']
min_distance = 25
peaks, _ = find_peaks(signal, distance=min_distance)
valleys, _ = find_peaks(-signal, distance=min_distance)

#plot picchi
fig1 = go.Figure()
fig1.add_trace(go.Scatter(x=list(peaks), y=signal[peaks], mode='markers', marker=dict(symbol='x',color='black'),name="valli"))
fig1.add_trace(go.Scatter(x=list(range(len(signal))), y=signal, mode='lines',name="accellerazione"))
fig1.update_layout(xaxis_title='tempo', yaxis_title='', title='Plot dei picchi con plantari')

fig1.show()

print("Indici dei picchi:", peaks)

mean_peak_value = np.mean(signal[peaks])
print('\n\nIl valore medio dei picchi è:',mean_peak_value)

#plot valli
fig2 = go.Figure()
fig2.add_trace(go.Scatter(x=list(valleys), y=signal[valleys], mode='markers', marker=dict(symbol='x',color='blue'),name="picchi"))
fig2.add_trace(go.Scatter(x=list(range(len(signal))), y=signal, mode='lines',name="accellerazione"))
fig2.update_layout(xaxis_title='tempo', yaxis_title=' ', title='Plot delle valli con plantari')

fig2.show()

print("Indici delle valli:", valleys)

mean_valleys_value = np.mean(signal[valleys])
print('\n\nIl valore medio delle valli è:',mean_valleys_value)
```

<h3> Calcolo le valli e i picchi della camminata senza plantari  </h3>

```python
signal = df_acce_without['x']
min_distance = 25
peaks_without, _ = find_peaks(signal, distance=min_distance)
valleys_without, _ = find_peaks(-signal, distance=min_distance)

#plot picchi
fig1 = go.Figure()
fig1.add_trace(go.Scatter(x=list(peaks_without), y=signal[peaks_without], mode='markers', marker=dict(symbol='x',color='black'),name="picchi"))
fig1.add_trace(go.Scatter(x=list(range(len(signal))), y=signal, mode='lines',name="accellerazione"))
fig1.update_layout(xaxis_title='tempo', yaxis_title='', title='Plot dei picchi senza plantari')

fig1.show()

print("Indici dei picchi:", peaks_without)

mean_peak_value_without = np.mean(signal[peaks_without])
print('\n\nIl valore medio dei picchi è:',mean_peak_value_without)

#plot valli
fig2 = go.Figure()
fig2.add_trace(go.Scatter(x=list(valleys_without), y=signal[valleys_without], mode='markers', marker=dict(symbol='x',color='blue'), name="valli"))
fig2.add_trace(go.Scatter(x=list(range(len(signal))), y=signal, mode='lines', name="accelerazione"))
fig2.update_layout(xaxis_title='tempo', yaxis_title=' ', title='Plot delle valli senza plantari')

fig2.show()

print("Indici delle valli:", valleys_without)

mean_valleys_value_without = np.mean(signal[valleys_without])
print('\n\nIl valore medio delle valli è:',mean_valleys_value_without)
```

<h3> Calcolo frequenze (tra cui frequenza media)</h3>

```python
# Calcolo la frequenza di campionamento con plantari
df_acce_with['time'] = pd.to_datetime(df_acce_with['time'], unit='ns')
time_diff = df_acce_with['time'].diff()
sampling_frequency = 1 / time_diff.mean().total_seconds()
print(f"\nFrequenza di campionamento con plantari: {sampling_frequency} Hz")

# Calcolo la frequenza di campionamento senza plantari
df_acce_without['time'] = pd.to_datetime(df_acce_without['time'], unit='ns')
time_diff_without = df_acce_without['time'].diff()
sampling_frequency_without = 1 / time_diff_without.mean().total_seconds()
print(f"\nFrequenza di campionamento senza plantari: {sampling_frequency_without} Hz")

# Calcolo la media delle frequenze di campionamento
mean_sampling_frequency = (sampling_frequency + sampling_frequency_without) / 2
print(f"\nFrequenza di campionamento media: {mean_sampling_frequency} Hz")
```

<h3> Calcolo medie e differenze con segno del segnale </h3>

```python
mean_x_with = df_acce_with['x'].mean()
mean_x_without = df_acce_without['x'].mean()


diff_with = df_acce_with['x'] - mean_x_with
diff_without = df_acce_without['x'] - mean_x_without


print(f"Media lungo l'asse x con plantari: {mean_x_with}\n")
print(f"Media lungo l'asse x senza plantari: {mean_x_without}\n\n")

print("Differenze con segno rispetto alla media con plantari:\n")
print(diff_with.head())

print("\n\nDifferenze con segno rispetto alla media senza plantari:\n")
print(diff_without.head())
```

<h3> Applico un filtro passa-basso </h3>

```python
def butter_lowpass(data, cutoff_freq, sampling_freq, order=3):
    nyquist = 0.5 * sampling_freq
    normal_cutoff = cutoff_freq / nyquist
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    y = lfilter(b, a, data)
    return y

cutoff_frequency = 8.0 

# Applico il filtro passa-basso ai dati con plantari
df_acce_with['x_filtered'] = butter_lowpass(df_acce_with['x'], cutoff_frequency, mean_sampling_frequency)

# Applico il filtro passa-basso ai dati senza plantari
df_acce_without['x_filtered'] = butter_lowpass(df_acce_without['x'], cutoff_frequency, mean_sampling_frequency)

selected_data_with = df_acce_with[start_index:end_index]
selected_data_without = df_acce_without[start_index:end_index]

fig_with = go.Figure()
fig_with.add_trace(go.Scatter(x=selected_data_with.index, y=selected_data_with['x'], name='Dati originali'))
fig_with.add_trace(go.Scatter(x=selected_data_with.index, y=selected_data_with['x_filtered'], name='Dati filtrati'))
fig_with.update_layout(title='Confronto dati con plantari prima e dopo filtro passa-basso')
fig_with.show()

fig_without = go.Figure()
fig_without.add_trace(go.Scatter(x=selected_data_without.index, y=selected_data_without['x'], name='Dati originali'))
fig_without.add_trace(go.Scatter(x=selected_data_without.index, y=selected_data_without['x_filtered'], name='Dati filtrati'))
fig_without.update_layout(title='Confronto dati senza plantari prima e dopo filtro passa-basso')
fig_without.show()
```

<h3> Calcolo medie e differenze con segno del segnale filtrato </h3>

```python
mean_x_with_filtered = df_acce_with['x_filtered'].mean()

mean_x_without_filtered = df_acce_without['x_filtered'].mean()

diff_with_filtered = df_acce_with['x_filtered'] - mean_x_with_filtered

diff_without_filtered = df_acce_without['x_filtered'] - mean_x_without_filtered

print(f"Media del segnale filtrato lungo l'asse x con plantari: {mean_x_with_filtered}\n")
print(f"Media del segnale filtrato lungo l'asse x senza plantari: {mean_x_without_filtered}\n\n")

print("Differenze con segno rispetto alla media con plantari per il segnale filtrato:\n")
print(diff_with_filtered.head())

print("\n\nDifferenze con segno rispetto alla media senza plantari per il segnale filtrato:\n")
print(diff_without_filtered.head())
```
